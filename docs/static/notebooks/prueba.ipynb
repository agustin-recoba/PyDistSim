```python
import sys
import os

sys.path.insert(0, os.path.abspath("../../../"))
from pymote import NetworkGenerator
from pymote import *
```

```python
from pymote.logger import logger

logger.setLevel(0)
```

```python
net_gen = NetworkGenerator(n_count=15)
net = net_gen.generate_random_network()
```

```python
net.show()
```

```python
from pymote.algorithms.ideal_broadcast_definition import Flood

# Asigna el algoritmo
net.algorithms = ((Flood, {"informationKey": "I"}),)

# Asigna el mensaje a enviar, la información inicial
some_node = net.nodes()[0]
some_node.memory["I"] = "Hello distributed world"
```

```python
from pymote.npickle import write_npickle, read_npickle

# Escribir la red + algoritmo en un archivo
write_npickle(net, "mreza1.tar.gz")
net_from_file = read_npickle("mreza1.tar.gz")
net_from_file
```

```python
# Correr la simulación
# %run ../../../pymote/gui/simulationgui.py
```

```python
sim = Simulation(net)
```

```python
sim.run()
```
