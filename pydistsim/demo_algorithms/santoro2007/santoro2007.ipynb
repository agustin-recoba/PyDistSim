```python
from pydistsim import Simulation, NetworkGenerator
from pydistsim.demo_algorithms.santoro2007.yoyo import YoYo
from pydistsim.network.rangenetwork import BidirectionalRangeNetwork

# Create a custom network to run YoYo on
# (Using PyDistSim's commRange atribute to create edges)
net = BidirectionalRangeNetwork()
sim = Simulation(net)

# Define node range and positions (IDs are dependent on position in list)
node_range = 101
node_positions = [
    (100, 500),
    (300, 400),
    (300, 100),
    (200, 500),
    (400, 200),
    (100, 400),
    (400, 100),
    (200, 400),
    (400, 300),
    (400, 400),
    (300, 200),
]

# Add nodes
for node_pos in node_positions:
    net.add_node(pos=node_pos, commRange=node_range)

# Show network
net.show()

# Set algorithm and create simulation
sim.algorithms = (YoYo,)

# Run simulation
sim.run()
```

```python
from networkx import Graph

# Create a custom network to run YoYo on
# (With custom edges)
# This network is the one used in the book
# Design and Analysis of Distributed Algorithms by Nicola Santoro
net = BidirectionalRangeNetwork()

# Define node range and positions (IDs are dependent on position in list)
node_range = 1
node_positions = [
    (350, 500),
    (150, 500),
    (100, 400),
    (50, 300),
    (250, 500),
    (450, 500),
    (300, 400),
    (200, 400),
    (350, 300),
    (400, 400),
    (150, 400),
    (250, 300),
    (450, 300),
    (300, 200),
    (125, 300),
    (150, 200),
    (200, 200),
    (300, 100),
    (200, 100),
]

# Define adjacent nodes
adj_dict = {
    0: [6, 8, 9],
    1: [2, 10, 7],
    2: [3, 14],
    3: [15],
    4: [7, 11, 6],
    5: [9, 12],
    6: [8],
    7: [11],
    9: [12],
    10: [14, 15],
    11: [16, 13],
    13: [18, 17],
    14: [15],
    16: [18, 17],
}

# Add nodes
for node_pos in node_positions:
    net.add_node(pos=node_pos, commRange=node_range)

# Get nodes list
nodes = list(net.nodes())

# Create edges
for node_1 in adj_dict:
    for node_2 in adj_dict[node_1]:
        net.add_edge(nodes[node_1], nodes[node_2])

# Show network
net.show()

# Set algorithm and create simulation
sim = Simulation(net)
sim.algorithms = (YoYo,)

# Run simulation
sim.run()
```
